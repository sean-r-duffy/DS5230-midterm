## Some needed terms
- Zillow Home Value Index (ZHVI): A measure of the typical home value and market changes across a given region and housing type. It reflects the typical value for homes in the 35th to 65th percentile range.
- Zillow Home Value Forecast (ZHVF): A month-ahead, quarter-ahead and year-ahead forecast of the Zillow Home Value Index (ZHVI). ZHVF is created using the all homes, mid-tier cut of ZHVI.
- Zillow Observed Rent Index (ZORI): A smoothed measure of the typical observed market rate rent across a given region. ZORI is a repeat-rent index that is weighted to the rental housing stock to ensure representativeness across the entire market, not just those homes currently listed for-rent. The index is dollar-denominated by computing the mean of listed rents that fall into the 40th to 60th percentile range for all homes and apartments in a given region, which is weighted to reflect the rental housing stock.

```python
median_ZORI_2018 = grouped_df.median().sort_values('ZORI 2018', ascending=False)
median_ZORI_2018 = median_ZORI_2018.dropna(subset=['ZORI 2018'])
sns.barplot(data= median_ZORI_2018, x= median_ZORI_2018.index, y='ZORI 2018', palette= orange_blue, alpha= 0.9)
plt.title('Median Rental Prices by State 2018')
plt.ylabel('Median ZORI Value ($)')
plt.xlabel('State')
plt.xticks(rotation=90, fontsize=8)
plt.grid(alpha=0.3)
plt.show()
print(median_ZORI_2018['ZORI 2018'].head(10))
```

```python
filtered_data = df.dropna(subset=['ZORI 2018', 'ZORI 2023'])
mean_zori_2018_filtered = filtered_data.groupby('State')['ZORI 2018'].mean()
mean_zori_2023_filtered = filtered_data.groupby('State')['ZORI 2023'].mean()
mean_zori_df_filtered = pd.DataFrame({'ZORI 2018': mean_zori_2018_filtered, 'ZORI 2023': mean_zori_2023_filtered}).reset_index()
fig, ax = plt.subplots(figsize=(15, 8))
bar_width = 0.4
index_filtered = mean_zori_df_filtered.index
bar1 = ax.bar(index_filtered, mean_zori_df_filtered['ZORI 2018'], bar_width, label='ZORI 2018', color='royalblue')
bar2 = ax.bar(index_filtered + bar_width, mean_zori_df_filtered['ZORI 2023'], bar_width, label='ZORI 2023', color='darkorange')
ax.set_xlabel('State')
ax.set_ylabel('Mean ZORI Value ($)')
ax.set_title('Mean ZORI Values by State for 2018 and 2023')
ax.set_xticks(index_filtered + bar_width / 2)
ax.set_xticklabels(mean_zori_df_filtered['State'])
ax.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
filtered_data = df.dropna(subset=['New Construction 2018', 'New Construction 2019', 'New Construction 2020', 'New Construction 2021', 'New Construction 2022', 'New Construction 2023'])
states = filtered_data['State'].unique()
n_states = len(states)

grid_size = int(np.ceil(np.sqrt(n_states)))
fig, axes = plt.subplots(grid_size, grid_size, figsize=(20, 20), sharex=True, sharey=True)
fig.suptitle('New Construction Trends 2018-2023 by State', fontsize=16)
axes = axes.flatten()
for i, state in enumerate(states):
    state_data = filtered_data[filtered_data['State'] == state]
    mean_new_construction = state_data[['New Construction 2018', 'New Construction 2019', 'New Construction 2020', 'New Construction 2021', 'New Construction 2022', 'New Construction 2023']].mean()
    axes[i].plot(mean_new_construction.index, mean_new_construction, marker='o', linestyle='-', color='royalblue')
    axes[i].set_title(state)
    axes[i].tick_params(labelrotation=45)
for i in range(n_states, len(axes)):
    axes[i].axis('off')
fig.text(0.5, 0.15, 'Years 2018-2023', ha='center', va='center', fontsize=12)
fig.text(0.02, 0.5, 'Avg. New Monthly Construction Projects', ha='center', va='center', rotation='vertical', fontsize=12)
plt.show()
```
